ALURA/Redes Neurais Convolucionais - Deep Learning com PyTorch/

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.signal import convolve

x = np.linspace(0, 100, 100)
y = 10 * np.sin(x) * np.random.rand(x.shape[0])

plt.figure(figsize=(12, 4))
plt.plot(x, y)
```

```python
def show(valores, title):
  plt.figure(figsize=(len(valores), 2))
  plt.imshow(valores[np.newaxis, :], cmap='gray')
  for k, s in enumerate(valores):
    plt.text(k, 0, '{:.1f}'.format(s), fontsize=16, color='red', ha='center', va='center')
  plt.title(title, fontsize=18)
  plt.yticks([])
```

```python
sinal = y[5:15]
show(sinal, 'Sinal')

plt.figure(figsize=(9.66, 5))
plt.plot(sinal)
```

```python
kernel = np.asarray([1.0, 0.0, -1.0])
show(kernel, 'Kernel')
```

```python
kernel_inv = np.flip(kernel)
show(kernel_inv, 'Kernel invertido')
```

```python
u = 2
deslocamento = [float('nan')] * u
kernel_desl = np.hstack((deslocamento, kernel_inv))

show(kernel_desl, 'Kernel deslocado')
show(sinal, 'Sinal')
```

```python
ativacao = convolve(sinal, kernel, mode='valid')
show(ativacao, 'Mapa de ativação')
```

```python
plt.figure(figsize=(12, 4))
plt.plot(sinal, color='k', linewidth=4)
plt.imshow(ativacao[np.newaxis, :], cmap='Reds', aspect='auto', alpha=0.8, extent=(0.5, 8.5, -10, 10))
```

```python
sinal2 = y[:50]
ativacao2 = convolve(sinal2, kernel, mode='valid')
ativacao2[ativacao2 < 0] = 0

plt.figure(figsize=(15, 4))
plt.plot(sinal2, color='k', linewidth=4)
plt.imshow(ativacao2[np.newaxis, :], cmap='Reds', aspect='auto', alpha=0.8, extent=(0.5, 48.5, -10, 10))
plt.xlim(-0.5, 50.5)
plt.colorbar()
```
