```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
def plotmodel(X, Y, w1, w2, b, p=None):
    fig = plt.figure(figsize=(10,8))
    plt.scatter(X[:,0], X[:,1], c=Y, edgecolor='k')
    plt.grid(True)
    
    xmin, xmax = plt.gca().get_xlim()
    ymin, ymax = plt.gca().get_ylim()
    
    plt.axhline(0, xmin, xmax, color='k', linewidth='1')
    plt.axvline(0, ymin, ymax, color='k', linewidth='1')
    
    x = np.linspace(xmin, xmax, 50)
    y = (-w1*x -b) / w2
    
    plt.plot(x,y)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    
    if p is not None:
        plt.plot(p[0], p[1], marker='*', markersize=20)
```

```python
np.random.seed(46)
X, Y = make_classification(n_features=2, n_redundant=0, n_clusters_per_class=1, n_informative=1)
```

```python
w1 = 5
w2 = 1
b = -0.4

torch.manual_seed(42)

perceptron = nn.Linear(in_features=2, out_features=1)
perceptron.weight = nn.Parameter(torch.Tensor([[w1, w2]]))
perceptron.bias = nn.Parameter(torch.Tensor([b]))

sigmoide = nn.Sigmoid()

# for name, tensor in perceptron.named_parameters():
#     print(name, tensor.data.numpy()

# (17, 21, 43, 66)

for k, idx in enumerate([17, 21, 43, 66]):
    print(k, idx)
```

```python
plotmodel(X, Y, w1, w2, b)

markers = ('^', 'v', '>', '<')
colors = ('r', 'g', 'b', 'gray')

for k, idx in enumerate([17, 21, 43, 66]):
    x = torch.Tensor(X[idx])
    
    ret = perceptron(x)
    act = sigmoide(ret)    
    
    label = 'ret: {:5.2f}'.format(ret.data.numpy()[0]) + ' act: {:5.2f}'.format(act.data.numpy()[0])
    plt.plot(x[0], x[1], marker=markers[k], color=colors[k], markersize=15, label=label)
 

# plt.show()
```
