## Importando o módulos de neural networks

```python
import torch
from torch import nn
```

## Definindo parâmetros do perceptron

```python
torch.manual_seed(42)

perceptron = nn.Linear(in_features=3, out_features=1)
print(perceptron)
```

## Checando valores iniciais dos pesos e bias

```python
for nome, tensor in perceptron.named_parameters():
    print(nome, tensor.data)
    
print('')
print(perceptron.weight.data)
print(perceptron.bias.data)
```

## Convertendo os valores dos tensores para valores numpy

```python
w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data.numpy()[0]

print(w1,w2,w3,b)

#w1 * x1 + w2 * x2 + w3 * x3 + b = 0
```

## Criando função para 'plotar' dimensões do perceptron

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot3d(perceptron):
    
    w1, w2, w3 = perceptron.weight.data.numpy()[0]
    b = perceptron.bias.data.numpy()[0]
                                     
    x1 = np.linspace(-1, 1, 10)
    x2 = np.linspace(-1, 1, 10)

    X1, X2 = np.meshgrid(x1, x2)
    
    X3 = (b - w1*X1 - w2*X2) / w3

    fig = plt.figure(figsize=(10, 8))
    
    ax = fig.gca(projection='3d')
    ax.view_init(azim=180)

    ax.plot_surface(X1, X2, X3, cmap='plasma')
    ax.set_xlabel('$X1$')
    ax.set_ylabel('$X2$')
    ax.set_zlabel('$X3$')
```

```python
plot3d(perceptron)
```

## Fazendo o 'forward'
#### usando um tensor (ponto) de entrada no perceptron , que gera um tensor de saída (representando a classificação linear

```python
X = torch.Tensor([0, -1, 2])
y = perceptron(X)

print(y)

plot3d(perceptron)
plt.plot([X[0]], [X[1]], [X[2]], color='r', marker='^', markersize=20)
```

#### Ao se mudar a coordenada do tensor de entrada para baixo do plano, percebe-se que o tensor de saída tem o sinal alterado

```python
X = torch.Tensor([0, 1, 2])
y = perceptron(X)

print(y)

plot3d(perceptron)
plt.plot([X[0]], [X[1]], [X[2]], color='r', marker='^', markersize=20)
```
